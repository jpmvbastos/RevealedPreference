```python
import pandas as pd
import numpy as np
from revealpref import revealpref
```

```python
# Example A: satisfies both WARP and SARP
# Each chosen bundle is the most expensive one in its own price period
# (own expenditure is the maximum in its row of the pq matrix), so the RP
# relation is a strict hierarchy: bundle 1 > bundle 2 > bundle 3. No cycles exist.
# Expected: 0.0% violate WARP, satisfies SARP

p = pd.DataFrame([[1,1],[1,3],[3,1]])
q = pd.DataFrame([[6,6],[4,2],[2,2]])

revealpref(p, q, axiom='both')
```

```python
# Example B: violates WARP (and therefore SARP)
# Bundles 2 and 5 are mutually directly revealed preferred to each other:
# at the prices faced in period 2, bundle 2 was chosen when bundle 5 was affordable,
# AND at the prices faced in period 5, bundle 5 was chosen when bundle 2 was affordable.
# Expected: 10.0% violate WARP (1 out of 10 pairs), does not satisfy SARP

p = pd.DataFrame([[4,8,10],[6,6,8],[5,5,6],[5,6,6],[6,5,8]])
q = pd.DataFrame([[7,4,4],[4,9,5],[6,10,6],[6,9,5],[4,9,5]])

revealpref(p, q, axiom='both')
```

```python
# Example C: passes WARP but fails SARP (from README)
# All three direct RP edges form a 3-cycle in the RP graph: 1→2→3→1.
# No single pair is mutually directly revealed preferred, so WARP passes.
# But detecting the SARP violation still requires the transitive closure:
# TC[1,3]=1 only via the indirect path 1→2→3, and TC[3,1]=1 from the direct
# edge 3→1 — together they form a cycle that only Floyd-Warshall surfaces.
# Expected: 0.0% violate WARP, does not satisfy SARP

p = pd.DataFrame([[2,2,2],[1,3,2],[2,1.5,5]])
q = pd.DataFrame([[2,2,2],[3,1,2],[4,1,1.5]])

revealpref(p, q, axiom='both')
```
